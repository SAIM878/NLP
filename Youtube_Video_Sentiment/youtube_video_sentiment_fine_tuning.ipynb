```python
import os
dir = "text/"
text_dir_contents = os.listdir(dir)
print("Contents of text directory:", text_dir_contents)

negative_dir = os.path.join("text/", "negative")
positive_dir = os.path.join("text/", "positive")
```

```python
import pandas as pd
def load_data_from_folder(folder_path, label):
    texts = []
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        if os.path.isfile(file_path):
            with open(file_path, 'r', encoding='utf-8') as file:
                text = file.read().strip()
                texts.append((text, label))
    return texts

negative_samples = load_data_from_folder(negative_dir, 0)
positive_samples = load_data_from_folder(positive_dir, 1)

all_samples = negative_samples + positive_samples
df = pd.DataFrame(all_samples, columns=["text", "label"])
df = df.sample(frac=1).reset_index(drop=True)
print(df.head())
```

```python
import nltk
import pandas as pd
from nltk.tokenize import sent_tokenize

# Download required resources for sentence tokenization
nltk.download('punkt')
```

```python

# Function to split text into sentences and label them
def split_into_sentences(text, label):
    sentences = sent_tokenize(text)
    return [(sentence, label) for sentence in sentences]

# Create a new list to hold sentence-level data
sentence_data = []

# Iterate over the dataframe and split texts into sentences
for index, row in df.iterrows():
    text = row['text']
    label = row['label']
    sentences = split_into_sentences(text, label)
    sentence_data.extend(sentences)

# Create a new DataFrame for sentence-level data
sentence_df = pd.DataFrame(sentence_data, columns=["sentence", "label"])

# Display the first few rows of the new DataFrame
print(sentence_df.head())
```

```python
combined_df = pd.concat([sentence_df, df_selected], ignore_index=True)
combined_df = combined_df.sample(frac=1).reset_index(drop=True)

print(combined_df.head())
```

```python
from datasets import Dataset
dataset = Dataset.from_pandas(combined_df)
```

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

def tokenize_function(examples):
    return tokenizer(examples['sentence'], padding='max_length', truncation=True)

# Convert DataFrame to Dataset
from datasets import Dataset

combined_dataset = Dataset.from_pandas(combined_df)

# Tokenize the dataset
tokenized_combined_dataset = combined_dataset.map(tokenize_function, batched=True)
```

```python
train_test_split = tokenized_combined_dataset.train_test_split(test_size=0.2)
train_dataset = train_test_split['train']
eval_dataset = train_test_split['test']

train_dataset.set_format('torch', columns=['input_ids', 'attention_mask', 'label'])
eval_dataset.set_format('torch', columns=['input_ids', 'attention_mask', 'label'])
```

```python
from transformers import BertForSequenceClassification, Trainer, TrainingArguments

# Define the model
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)

# Define training arguments with increased batch sizes and mixed precision training
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=32,  # Further increased batch size
    per_device_eval_batch_size=32,   # Further increased batch size
    evaluation_strategy='epoch',
    save_strategy='epoch',
    learning_rate=2e-5,
    logging_dir='./logs',
    logging_steps=10,
    fp16=True,  # Enable mixed precision training
)

# Define the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
)

# Start training
trainer.train()
```

```python
model.save_pretrained('./fine_tuned_bert')
tokenizer.save_pretrained('./fine_tuned_bert')
```

```python
results = trainer.evaluate()
print(f"Evaluation results: {results}")
```

```python
from transformers import BertTokenizer, BertForSequenceClassification, AutoConfig
import torch

# Define the directory containing the model and tokenizer files
model_dir = '.'

# Load the configuration, model, and tokenizer
config = AutoConfig.from_pretrained(model_dir)
model = BertForSequenceClassification.from_pretrained(model_dir, config=config)
tokenizer = BertTokenizer.from_pretrained(model_dir)

# Function to make predictions on a given text
def predict_sentiment(text):
    # Tokenize the input text
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=512)

    # Set the model to evaluation mode
    model.eval()

    # Disable gradient calculation for faster inference
    with torch.no_grad():
        # Make the prediction
        outputs = model(**inputs)
        logits = outputs.logits
        predicted_class = torch.argmax(logits, dim=1).item()

    # Convert predicted class to label (0 or 1)
    label = 'negative' if predicted_class == 0 else 'positive'

    return label

# Example usage
text = "I hate this product! It works bad and is easy to use."
prediction = predict_sentiment(text)
print(f"The sentiment of the text is: {prediction}")
```
